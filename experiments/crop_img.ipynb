```python
from pathlib import Path
from PIL import Image
import pandas as pd
import numpy as np
```

```python
img_folder = Path('../data_round_2/final')
filename_to_path = {x.name: x for x in img_folder.glob('**/*.jpeg')}
```

```python
val_annotations_csv = "../data_round_2/best_model_val_data_yolo_annotations.csv"
train_annotations_csv = "../data_round_2/best_model_train_data_yolo_annotations.csv"
train_df = pd.read_csv(train_annotations_csv)
val_df = pd.read_csv(val_annotations_csv)
```

```python
train_df
```

```python
IMG_SCALED_DIM = 224
PADDING = 20
def extract_boxes_from_image(df, output):
    for index, row in df.iterrows():
        img_filename = row["img_fName"].split('.')[0]
        print(img_filename)
        try:
            img = Image.open(filename_to_path[img_filename + ".jpeg"])
            #img = Image.open("../data_round_2/final/train_00000.jpeg")
            #img = ImageOps.exif_transpose(img)
        except:
            print("fail to open")
            return

        i = 0
        while True:
            output_img_filename = output / f'{img_filename}_{i:03d}.jpeg'
            i = i + 1
            print(output_img_filename)
            if not output_img_filename.exists():
                break

        f_name, w, h, x_tl, y_tl, x_br, y_br, label = row
        # define the bounding box (left, upper, right, lower)
        bbox = (int(x_tl), int(y_tl), int(x_br), int(y_br))
        bbox = [bbox[0]-PADDING, bbox[1] - PADDING, bbox[2] + PADDING, bbox[3] + PADDING]
        cropped_img = img.crop(bbox)
        cropped_img.save(output_img_filename, quality=100, subsampling=0)
```

```python
output_train = img_folder = Path('../data_round_2/cropped/train')
output_train.mkdir(exist_ok=True)
extract_boxes_from_image(train_df, output_train)
```

```python
output_val = img_folder = Path('../data_round_2/cropped/val')
output_val.mkdir(exist_ok=True)
extract_boxes_from_image(val_df, output_val)
```
