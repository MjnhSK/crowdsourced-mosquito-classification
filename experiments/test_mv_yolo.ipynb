```python
# from python notebook cells, get all the img names listed in data_round_2/closedSet/train.csv and /val.csv, and then move all the images with corresponding name in data_yolo/image to a new directory
# purpose is to prepare the images for yolo training

import os
import shutil
import pandas as pd


def get_img_names_from_csv(csv_path):
    df = pd.read_csv(csv_path)
    img_names = df['img_fName'].tolist()
    return img_names
```

```python
train = pd.read_csv('../data_round_2/closedSet/train.csv')
train.shape
```

```python
# get all img names from train.csv and val.csv
train_img_names = get_img_names_from_csv('../data_round_2/closedSet/train.csv')
val_img_names = get_img_names_from_csv('../data_round_2/closedSet/val.csv')
```

```python
train_img_names
```

```python
# move all images with corresponding name in data_yolo/image to a new directory
img_dir = '../data_yolo_new/image/train'
os.makedirs(img_dir, exist_ok=True)
for img_name in train_img_names:
    shutil.copy(img_name, img_dir)
```
